```python
# load needed python libraries matplotlib, numpy, and scikit-rf
import matplotlib.pyplot as plt
import numpy as np
import skrf as rf
import json
```

```python
solt = rf.read_all('pcb_SOLT_warm/')
fghz = solt['short'].f/1e9
```

```python
plt.plot(fghz,solt['short'].s_db[:,0,0],color="red")
plt.plot(fghz,solt['open'].s_db[:,0,0],color="green")
plt.plot(fghz,solt['load'].s_db[:,0,0],color="blue")
plt.plot(fghz,solt['thru'].s_db[:,0,0],color="purple")
plt.grid()
plt.xlim(4,8)
plt.ylim(-40,10)
plt.ylabel('$S_{11} [dB]$')
plt.xlabel('Frequency [GHz]')
plt.legend(['short','open','load','thru'])
```

```python
plt.plot(fghz,solt['short'].s_db[:,1,0],color="red")
plt.plot(fghz,solt['open'].s_db[:,1,0],color="green")
plt.plot(fghz,solt['load'].s_db[:,1,0],color="blue")
plt.plot(fghz,solt['thru'].s_db[:,1,0],color="purple")
plt.grid()
plt.xlim(4,8)
plt.ylim(-40,10)
plt.ylabel('$S_{21} [dB]$')
plt.xlabel('Frequency [GHz]')
plt.legend(['short','open','load','thru'])
```

```python
plt.plot(fghz,(solt['short'].s_deg_unwrap[:,0,0] - solt['open'].s_deg_unwrap[:,0,0])%360,color="red")
plt.grid()
plt.xlim(4,8)
plt.ylim(0,360)
plt.ylabel('$S_{11} (short - open) [deg]$')
plt.xlabel('Frequency [GHz]')
```

```python
plt.plot(fghz,solt['thru'].s_db[:,1,0],color="blue")
plt.plot(fghz,solt['thru'].s_db[:,0,1],color="purple")
plt.grid()
plt.xlim(4,8)
plt.ylim(-40,10)
plt.ylabel('$S_{xx} [dB]$')
plt.xlabel('Frequency [GHz]')
plt.legend(['$S_{21}$ (thru)','$S_{12}$ (thru)'])
```
